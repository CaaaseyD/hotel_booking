### Hotel Bookings Data Cleansing

```python
import pandas as pd
import numpy as np

missing_values = ["undefined", " ", "none", "-" ]
data_path = "hotel_bookings.csv"
hotel = pd.read_csv(data_path)

#hotel.head
#hotel.shape
#hotel.info

"""
Convert column types
1. create a new column to combine arrival day, month and year as a datetime type.
2. convert current column reservation_status_date to a datetime type.
"""
hotel['arrival_date'] = pd.to_datetime(hotel['arrival_date_year'].astype(str) + '/' + hotel['arrival_date_month'].astype(str) + '/' + hotel['arrival_date_day_of_month'].astype(str))
hotel['reservation_status_date'] = pd.to_datetime(hotel['reservation_status_date'])

"""
Removing the unwanted columns
1. if more than 70% of values in a column are missing and there is no way to fill in the missing values, 
then the column can be dropped completely from the dataset.
2. drop duplicated / same value columns
"""
# Finding the number of missing values in every column
np.sum(hotel.isnull())

# remove columns missing value counts >= 70% total counts
for col in hotel.columns:
    if np.sum(hotel[col].isnull()) >= (hotel.shape[0] * 0.7):
        hotel.drop(columns=col, inplace=True, axis=1)

hotel.drop(columns=['arrival_date_year', 'arrival_date_month', 'arrival_date_day_of_month', 'arrival_date_week_number'], inplace=True, axis=1)

hotel.info

```
